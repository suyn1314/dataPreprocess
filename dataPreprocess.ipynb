# Step 6 Use matplotlib show 2D images about data.
[Matplotlib](https://matplotlib.org/)
### Plot the distributions of two classes, 1. GDP < 58000 and  2. GDP >= 58000, of COVID-19 State Data using different colors and different  markers where x-axis is the Pollution and y-axis the Mortality-rate.

```python
import matplotlib.pyplot as plt 

df_GDP = df[df["GDP"] >= 58000]
pollution_column = list(df_GDP['Pollution'])
mortality_GDP = list (df_GDP['Deaths']/df_GDP['Infected'])
mortality_rate = [round(num, 4) for num in mortality_GDP]
GDP_x = sorted(pollution_column)
GDP_y = sorted(mortality_rate)

df_GDP_low = df[df["GDP"] < 58000]
pollution_column_low = list(df_GDP_low['Pollution'])
mortality_GDP_low = list (df_GDP_low['Deaths']/df_GDP_low['Infected'])
mortality_rate_low = [round(num, 4) for num in mortality_GDP_low]
GDP_x_low = sorted(pollution_column_low)
GDP_y_low = sorted(mortality_rate_low)

fig = plt.figure()
fig.patch.set_facecolor('xkcd:white')

plt.title("COVID-19 GDP") 
plt.xlabel("x-axis is the Pollution") 
plt.ylabel("y-axis the Mortality-rate") 

plt.plot(GDP_x,GDP_y)
plt.plot(GDP_x_low,GDP_y_low,color=(255/255,100/255,100/255))
plt.show()
```

# Step 7
### Show the proportion of following three classes of COVID-19 State Data using pie chart.

**<font color = blue size=3> About the class: </font>** <p>
 1、 Mortality-rate < 0.02 <p>
 2、 Mortality-rate between 0.02 and 0.03 <p>
 3、Mortality-rate > 0.03 <p>

```python
mortality = list (df['Deaths']/df['Infected'])
mortality_rate = [round(num, 4) for num in mortality]

mr_2 = np.sum(list(map(lambda x: x < 0.02,mortality_rate )))
mr = np.sum(list(map(lambda x: 0.02 <= x <= 0.03,mortality_rate )))
mr_3 =  np.sum(list(map(lambda x: x > 0.02,mortality_rate )))

expend = [mr_2, mr, mr_3]
category = ['Mortality-rate < 0.02', 'Mortality-rate between 0.02 and 0.03', 'Mortality-rate > 0.03']
fig = plt.figure()
fig.patch.set_facecolor('xkcd:white')
plt.pie(expend, 
        labels = category,
       autopct='%1.1f%%')
plt.axis('equal')
plt.show
```
